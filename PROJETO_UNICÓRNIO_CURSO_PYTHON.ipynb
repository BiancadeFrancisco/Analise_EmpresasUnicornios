```python
df.columns
```

```python
# ALTERAR TIPOS DE DADOS:

df['Data_Adesao'] = pd.to_datetime(df['Data_Adesao'])

df['Data_Adesao'].head()
```

```python
# CRIAR COLUNA PARA MÊS E COLUNA PARA ANO:

df['Mes'] = pd.DatetimeIndex(df['Data_Adesao']).month
df['Ano'] = pd.DatetimeIndex(df['Data_Adesao']).year
```

```python
df.head()
```

```python
# FAZER TABELA ANALITICA:

Analise_Agrupada = df.groupby( by=['País', 'Ano', 'Mes', 'Empresa'] ).count()['ID'].reset_index()

Analise_Agrupada
```

```python
Analise_Agrupada.loc[
  Analise_Agrupada['País'] == 'Brazil'
]
```

```python
# AJUSTAR COLUNA VALOR:

df['Valor'] = pd.to_numeric(df['Valor'].apply(lambda linha: linha.replace('$', '')))

# pd.to_numeric = altera o tipo da coluna para numerica
# apply(lambda linha: linha.replace('$', '')) = aplicou uma função lambda que passou de linha em linha na coluna valor e alterou o caracter '$' para ''(nada).
```

```python
df.head()
```

```python
# TABELA ANALÍTICA PARA VISUALIZAR O VALOR DE CADA PAÍS:

Analise_Pais = df.groupby( by=['País'] ).sum()['Valor'].reset_index()

Analise_Pais.head()
```

```python
Analise_Valor = Analise_Pais.sort_values('Valor', ascending=False)
Analise_Valor.head(10)
```

```python
plt.figure( figsize=(15,6) )
plt.plot( Analise_Valor['País'], Analise_Valor['Valor'] )
plt.title('Analise do Valor por Pais')
plt.xticks( rotation=45, ha='right');
```
